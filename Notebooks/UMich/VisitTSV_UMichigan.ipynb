```python
import pandas as pd
import numpy as np
```

### Creating the visit TSV for UMichigan:

```python
directory1 = '/content/drive/MyDrive/powertools_umich_20230113.csv' #powertools output file for UMichigan batch 1
directory2 = '/content/drive/MyDrive/all_covid_details_umichigan.csv' #all covid admission details provided by UMichigan
```

```python
df1 = pd.read_csv(directory1)
df2 = pd.read_csv(directory2)
```

#### Grabbing all the cases in batch 1 using the powertools output file:

```python
cases_b1 = df1.drop_duplicates(subset ='case_ids')
cases_b1
```

#### Now let's locate the covid19-related admission details for this cohort:

```python
admission_details_b1 = df2.loc[df2.anon_mrn.isin(cases_b1['case_ids'])]
admission_details_b1
```

#### We only want details for patients that have been admitted to the hospital:

```python
admission_details_b1 = admission_details_b1.loc[admission_details_b1['covid19_hospital_admission']=='Yes']
```

```python
admission_details_b1
```

#### There are 211 patient admissions. Let's clean this table up a bit:

```python
admission_details_b1 = admission_details_b1[['anon_mrn', 'anon_date_covid19_hospital_admission', 'anon_date_covid19_hospital_discharge', 'covid19_discharge_type', 'anon_date_covid19_icu_admission', 'anon_date_covid19_icu_discharge']]
```

```python
admission_details_b1.rename(columns={'anon_mrn': 'case_ids', 'anon_date_covid19_hospital_admission': 'days_to_hospital_admission',
                                     'anon_date_covid19_hospital_discharge': 'days_to_hospital_discharge',
                                     'covid19_discharge_type':'discharge_type', 'anon_date_covid19_icu_admission': 'days_to_icu_admission',
                                     'anon_date_covid19_icu_discharge': 'days_to_icu_discharge'}, inplace=True)
```

```python
admission_details_b1['cases'] = admission_details_b1['case_ids']
```

```python
admission_details_b1
```

#### Looks good. Making sure that the discharge type property doesn't contain any potential PHI:

```python
admission_details_b1['discharge_type'].unique()
```

#### Now let's create the unique submitter ids:

```python
mrns = admission_details_b1['case_ids'].tolist()
print(mrns)
```

```python
count = 1
for i in range(len(mrns)):
  if i !=(len(mrns)-1):
    next = mrns[i+1]
  else:
    next = 'end'
  current = mrns[i]
  mrns[i] = mrns[i] + "_0" + str(count)
  if current == next:
    count = count+1
  else:
    count = 1
```

```python
admission_details_b1 = admission_details_b1.assign(submitter_id=mrns)
admission_details_b1
```

```python
admission_details_b1["submitter_id"] =   "Visit_" + admission_details_b1["submitter_id"]
```

```python
admission_details_b1
```

#### Everything looks good. Let's double check that we have 211 unique submitter ids:

```python
admission_details_b1.drop_duplicates(subset = 'submitter_id')
```

#### Visit node for UMichigan batch 1 is now complete. Let's download the file:
