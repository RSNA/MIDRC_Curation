```python
import pandas as pd
import numpy as np
```

### Creating the imaging TSV for UMichigan batch 1:

```python
directory1 = '/content/drive/MyDrive/powertools_umich_20230113.csv' #powertools output file for UMichigan batch 1
directory2 = '/content/all_imaging_studies_including_385_missing.csv' #all imaging study details provided by UMichigan and generated by me
```

```python
df1 = pd.read_csv(directory1)
df2 = pd.read_csv(directory2)
```

#### Grabbing all the studies in batch 1 using the powertools output file:

```python
studies_b1 = df1.drop_duplicates(subset='study_uid')
```

```python
studies_b1
```

#### There are 4932 studies in batch 1. Let's clean up this table a bit:

```python
studies_b1 = studies_b1[['case_ids', 'body_part_examined', 'modality', 'study_uid', 'study_description']]
```

```python
studies_b1.rename(columns={'modality': 'study_modality'}, inplace=True)
```

```python
studies_b1['submitter_id'] = studies_b1['study_uid']
studies_b1['cases'] = studies_b1['case_ids']
studies_b1['study_year_shifted'] = False
```

```python
studies_b1
```

#### Now that the table is cleaned up some, let's use the imaging details submitted by UMichigan (and also generated by me)to get more info:

```python
df2
```

#### Locating all the details for the studies in our batch. We're looking for the age at imaging, the days_to_study values, and the study year:

```python
study_details_b1 = df2.loc[df2.anon_study_uid.isin(studies_b1['study_uid'])]
```

```python
study_details_b1 = study_details_b1[['age','anon_study_uid','anon_study_date', 'study_year']]
```

```python
study_details_b1.rename(columns={'age': 'age_at_imaging', 'anon_study_date': 'days_to_study','anon_study_uid':'study_uid'}, inplace=True)
```

```python
study_details_b1
```

#### Let's merge these details with the studies_b1 table:

```python
studies_b1 = pd.merge(studies_b1, study_details_b1,  how='left', on=['study_uid'])
```

```python
studies_b1
```

#### Looks great. Let's now make sure the properties align with the MIDRC data model:

```python
studies_b1['body_part_examined'].unique()
```

```python
studies_b1['study_modality'].unique()
```

```python
studies_b1['study_description'].unique()
```

```python
studies_b1['study_year'].unique()
```

```python
studies_b1['age_at_imaging'].unique()
```

#### There are some patients in this cohort that are 90 years old. Let's remove the values above 89 and add the 'age_at_imaging_gt89' column to account for them:

```python
studies_b1.loc[studies_b1['age_at_imaging'] >= 90, 'age_at_imaging_gt89'] = 'Yes'
```

```python
studies_b1['age_at_imaging_gt89'].fillna('No', inplace = True)
```

```python
studies_b1.loc[studies_b1['age_at_imaging'] >= 90, 'age_at_imaging'] = np.NaN
```

#### Okay, we're almost done. Let's double check that the submitter ids are unique and that there are 4932 of them:

```python
studies_b1.drop_duplicates(subset='submitter_id')
```

#### Now the imaging study node for UMichigan batch 1 is ready and can be downloaded:

```python
studies_b1.to_csv('imaging_study_RSNA_20230117.csv', index = False)
```
