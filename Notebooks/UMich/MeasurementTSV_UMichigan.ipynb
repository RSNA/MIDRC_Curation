```python
import pandas as pd
import numpy as np
```

### Creating the measurement TSV for UMichigan batch 1:

```python
directory1 = '/content/drive/MyDrive/powertools_umich_20230113.csv' #powertools output file for UMichigan batch 1
directory2 = '/content/drive/MyDrive/umichigan_measurements_all_patients.csv' #all covid19 measurements provided by UMichigan
```

```python
df1 = pd.read_csv(directory1)
df2 = pd.read_csv(directory2)
```

#### Grabbing all the cases in batch 1 using the powertools output file:

```python
cases_b1 = df1.drop_duplicates(subset = 'case_ids')
cases_b1
```

#### There are 2270 cases in batch 1. Let's find all the covid19 measurements for this cohort:

```python
df2
```

```python
measurements_b1 = df2.loc[df2.anon_mrn.isin(cases_b1['case_ids'])]
measurements_b1
```

#### There are 7349 measurements for this cohort. Let's clean up the table a bit:

```python
measurements_b1.rename(columns={'anon_mrn': 'case_ids', 'anon_test_date': 'test_days_from_index', 'test_result': 'test_result_text'}, inplace=True)
```

```python
measurements_b1['test_name']= 'COVID-19'
measurements_b1['cases'] = measurements_b1['case_ids']
measurements_b1
```

```python
measurements_b1 = measurements_b1.sort_values(by=['case_ids', 'test_days_from_index'])
measurements_b1
```

#### After cleaning up the table some, let's make sure that the properites only have the accepted MIDRC data model values:

```python
measurements_b1['test_method'].unique()
```

```python
measurements_b1['test_result_text'].unique()
```

#### Let's create the unique submitter id for all the measurements:

```python
mrns = measurements_b1['case_ids'].tolist()
print(mrns)
```

```python
count = 1
for i in range(len(mrns)):
  if i !=(len(mrns)-1):
    next = mrns[i+1]
  else:
    next = 'end'
  current = mrns[i]
  mrns[i] = mrns[i] + "_0" + str(count)
  if current == next:
    count = count+1
  else:
    count = 1
```

```python
measurements_b1 = measurements_b1.assign(submitter_id=mrns)
measurements_b1
```

```python
measurements_b1["submitter_id"] =   "Measurement_" + measurements_b1["submitter_id"]
```

```python
measurements_b1
```

#### Looks good. Now let's make sure that the submitter ids are unique. There should be 7349 unique submitter ids:

```python
measurements_b1.drop_duplicates(subset = 'submitter_id')
```

#### Also let's make sure that there's at least one measurement per patient in the cohort:

```python
measurements_b1.drop_duplicates(subset = 'cases')
```

#### Everything checks out. The measurement table can now be downloaded:

```python
measurements_b1.to_csv('measurement_RSNA_20230117.csv', index = False)
```
