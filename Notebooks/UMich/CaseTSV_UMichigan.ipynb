```python
import pandas as pd
import numpy as np
```

### Creating the case TSV for UMichigan batch 1:



```python
directory1 = '/content/drive/MyDrive/powertools_umich_20230113.csv' #powertools output file for UMichigan batch 1
directory2 = '/content/drive/MyDrive/all_patients_umichigan.csv' #all patient data provided by UMichigan
directory3 = '/content/drive/MyDrive/all_covid_details_umichigan.csv' #all covid admission details provided by UMichigan
directory4 = '/content/drive/MyDrive/umichigan_measurements_all_patients.csv' #all covid19 measurements provided by UMichigan
directory5 = '/content/drive/MyDrive/all_imaging_studies_including_385_missing.csv' #all imaging study details provided by UMichigan and generated by me
```

```python
df1 = pd.read_csv(directory1)
df2 = pd.read_csv(directory2)
df3 = pd.read_csv(directory3)
df4 = pd.read_csv(directory4)
df5 = pd.read_csv(directory5)
```

#### Grabbing all the cases in batch 1 using the powertools output file:

```python
cases_b1 = df1.drop_duplicates(subset ='case_ids')
cases_b1
```

#### There are 2270 patients in batch 1. Locating all the patient detials for this cohort:

```python
cases_b1 = df2.loc[df2.anon_mrn.isin(cases_b1['case_ids'])]
```

```python
cases_b1
```

#### Let's clean this table up a bit:

```python
cases_b1.rename(columns={'anon_mrn': 'case_ids'}, inplace=True)
```

#### Dropping age because it's empty in this table. We'll find the age for each patient later:

```python
cases_b1.drop('age', axis=1, inplace=True)
```

#### Adding some more propoerties:

```python
cases_b1['site_id']= '302028'
cases_b1['submitter_id'] = cases_b1['case_ids']
cases_b1['gen3_linked_subjects_available'] = False
cases_b1['index_event'] = 'First COVID test'
cases_b1
```

```python
cases_b1['race'] = cases_b1['race'].map({'Black or African American': 'Black or African American',
                                                   'White or Caucasian': 'White',
                                                   'Unknown': 'Not Reported',
                                                   'Asian': 'Asian',
                                                   'Other': 'Other',
                                                   'White or Caucasian, Black or African American' : 'Other',
                                                   'Choose not to disclose': 'Not Reported',
                                                   'White or Caucasian, Other': 'Other',
                                                   'Asian, Unknown, White or Caucasian': 'Other',
                                                   'Native Hawaiian and Other Pacific Islander': 'Native Hawaiian or other Pacific Islander',
                                                   'Black or African American, Unknown': 'Other',
                                                   'Filipino': 'Asian',
                                                   'Not Reported': 'Not Reported',
                                                   'Other, White or Caucasian': 'Other',
                                                   'Black or African American, White or Caucasian': 'Other',
                                                   'American Indian and Alaska Native': 'American Indian or Alaska Native',
                                                   'Black or African American, Other': 'Other',
                                                   'Black or African American, American Indian and Alaska Native': 'Other',
                                                   'Chinese, White or Caucasian': 'Other',
                                                   'Asian, White or Caucasian': 'Other',
                                                   'Japanese, White or Caucasian': 'Other',
                                                   'Asian, Black or African American, White or Caucasian': 'Other',
                                                   'Filipino, Black or African American, White or Caucasian': 'Other',
                                                   'White or Caucasian, Asian': 'Other',
                                                   'White or Caucasian, American Indian and Alaska Native': 'Other',
                                                   'Chinese': 'Asian',
                                                   'Asian Indian': 'Asian',
                                                   'Unknown, White or Caucasian': 'Other',
                                                   'American Indian and Alaska Native, White or Caucasian': 'Other',
                                                   'Asian, Other': 'Other',
                                                   'American Indian and Alaska Native, Black or African American': 'Other'})
```

```python
cases_b1['race'].unique()
```

#### Let's determine which patients were covid19 positive so we can populate the 'covid19_positive' property. Locating all covid19 measurements for the cohort:

```python
measurements_b1 = df4.loc[df4.anon_mrn.isin(cases_b1['case_ids'])]
```

```python
measurements_b1
```

```python
measurements_b1.loc[measurements_b1['test_result'] == 'Positive', 'covid19_positive'] = 'Yes'
```

```python
covid_positive = measurements_b1.loc[measurements_b1['covid19_positive']=='Yes']
```

```python
covid_positive = covid_positive.drop_duplicates(subset = 'anon_mrn')
```

```python
covid_positive = covid_positive[['anon_mrn', 'covid19_positive']]
covid_positive
```

```python
covid_positive.rename(columns={'anon_mrn': 'case_ids'}, inplace=True)
```

#### Now that we have a list of the patients within the cohort that had at least one positive covid19 measurement, let's merge with our existing table:

```python
cases_b1 = pd.merge(cases_b1, covid_positive,  how='left', on=['case_ids'])
```

#### All the other patients are not covid positive, so we need to fill in 'No' for those patients:

```python
cases_b1['covid19_positive'].fillna('No', inplace = True)
```

#### Now we have to find out if the patient was ever in the icu or on a ventilator. Let's use the covid admission details table to find all the admission details for this cohort:

```python
admission_details_b1 = df3.loc[df3.anon_mrn.isin(cases_b1['case_ids'])]
```

```python
admission_details_b1 = admission_details_b1.drop_duplicates(subset='anon_mrn')
admission_details_b1
```

First let's locate the patients that were admitted to the icu:

```python
icu_indic = admission_details_b1[['anon_mrn', 'covid19_icu_admission']]
```

```python
icu_indic.rename(columns={'anon_mrn': 'case_ids', 'covid19_icu_admission':'icu_indicator'}, inplace=True)
```

#### Let's merge the icu data with our exisitng 'cases_b1' table:

```python
cases_b1 = pd.merge(cases_b1, icu_indic,  how='left', on=['case_ids'])
```

```python
cases_b1['icu_indicator'].unique()
```

#### Map the icu_indicator values to boolean:

```python
cases_b1['icu_indicator'] = cases_b1['icu_indicator'].map({'No': False, 'Yes': True})
```

```python
cases_b1
```

#### We don't have any patients that have ventilator information, so we won't populate that property:

```python
admission_details_b1['covid19_breathing_support_type'].unique()
```

```python
cases_b1['ventilator_indicator'] = np.nan
cases_b1['country_of_origin'] = np.nan
```

We also don't have info on the patients' country of origin, so that property is also left blank.

```python
cases_b1
```

#### Okay we're almost done. We need to get patient age. Let's use the imaging details table to grab the patient age:

```python
df5 = df5.sort_values(by=['anon_mrn', 'age'])
```

```python
df5 = df5.drop_duplicates(subset ='anon_mrn')
```

```python
df5 = df5.loc[df5.anon_mrn.isin(cases_b1['case_ids'])]
```

```python
df5 = df5[['anon_mrn', 'age']]
```

```python
df5.rename(columns={'anon_mrn': 'case_ids', 'age': 'age_at_index'}, inplace=True)
```

```python
df5
```

#### Now that we have the age for each patient in the cohort, let's merge with the 'cases_b1' table:

```python
cases_b1 = pd.merge(cases_b1, df5,  how='left', on=['case_ids'])
```

```python
cases_b1
```

```python
cases_b1['age_at_index'].unique()
```

#### There are some patients in this cohort that are 90 years old. Let's remove the values above 89 and add the 'age_at_index_gt89' column to account for them:

```python
cases_b1.loc[cases_b1['age_at_index'] >= 90, 'age_at_index_gt89'] = 'Yes'
```

```python
cases_b1['age_at_index_gt89'].fillna('No', inplace = True)
```

```python
cases_b1.loc[cases_b1['age_at_index'] >= 90, 'age_at_index'] = np.NaN
```

```python
cases_b1
```

#### Everything looks good. Let's double check that the submitter ids are unique. There should be 2270:

```python
cases_b1.drop_duplicates(subset='submitter_id')
```
