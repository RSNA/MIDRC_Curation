### Condensing clinical files from UCSF for the use in creating the Case Node, Measurement Node, and Imaging Study Node

```python
import pandas as pd
import numpy as np
```

##### Grab UCSF clinical files from Wyatt (most recent 'transferred exams' and 'all covid test results' csv files). The 'all covid results' file holds the info we need for the Measurement node. The 'transferred exams' file holds the info we need for the Case node and some of the Imaging Study node. Also grab a copy of the verbose report from POSDA for your batch.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
directory1 = '/content/drive/MyDrive/transferred exams - batch 26.csv'
directory2 = '/content/drive/MyDrive/all covid test results - batch 26.csv'
directory3 = '/content/drive/MyDrive/verbose_report_batch_15.csv'
```

##### Convert files to pandas dataframes:

```python
df = pd.read_csv(directory3,encoding='latin1')
df_all_covid = pd.read_csv(directory2)
df_transfer = pd.read_csv(directory1)
```

##### Grab all unique study uids:

```python
df = df.drop_duplicates(subset=['study_instance_uid'])
df
```

##### Remove unecessary columns from dataframe:

```python
df = df.drop(columns=['collection', 'site', 'patient_age', 'study_date', 'study_description', 'series_instance_uid', 'series_date', 'series_description', 'dicom_file_type','modality', 'num_files'])
```

```python
df_transfer
```

##### Merge transferred exam information with batch data:

```python
df_leftover = df.merge(df_transfer, left_on='study_instance_uid',right_on='Anon Study UID', how='inner')
```

```python
df_leftover
```

##### Now we have one of the input files ready with the study/case information:

```python
df_leftover.to_csv('studies_batch_16_clinical.csv', index = False)
```

##### Now let's create the input file that will hold the covid measurements:

```python
df_all_covid
```

```python
df_all_covid.rename(columns={'Anon MRN':'Anon_MRN'}, inplace=True)
```

##### Locate all the cases we need in df_all_covid:

```python
cases_needed = df_all_covid.loc[df_all_covid.Anon_MRN.isin(df_leftover['Anon MRN'])]
```

```python
df_leftover.rename(columns={'Anon MRN':'Anon_MRN'}, inplace=True)
```

```python
cases_needed
```

```python
cases_all_covid = cases_needed.drop_duplicates(subset=['Anon_MRN'])
cases_all_covid = cases_all_covid['Anon_MRN']
```
