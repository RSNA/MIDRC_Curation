```python
#Tom OSullivan

import pandas as pd
import pandas_bokeh
import matplotlib.pyplot as plt
import pgeocode
import geopandas as gpd
from shapely.geometry import Point
from geopandas import GeoDataFrame
pandas_bokeh.output_notebook()
import plotly.io as pio
pio.renderers.default = "notebook_connected"
import plotly.graph_objects as go

nomi = pgeocode.Nominatim('us')

edf = pd.read_csv('Midrc_Anon_Zips1.csv', usecols = ['zipcode','count'])
#edf = pd.read_csv('C:\Users\tosullivan\Documents\Midrc_Anon_Zips.csv', usecols = ['zipcode','count'])

# Check the data type of the 'zipcode' column
zipcode_dtype = edf['zipcode'].dtype

# If the data type is not already a string, you can convert it
if zipcode_dtype != 'str':
    edf['zipcode'] = edf['zipcode'].astype(str)
    print("Converted 'zipcode' column to string data type.")

# Print the data type of the 'zipcode' column after conversion
print(f"Data type of 'zipcode' column: {edf['zipcode'].dtype}")

# Check for missing values in the 'zipcode' column
missing_values = edf['zipcode'].isnull().sum()

# Print the number of missing values
print(f"Number of missing values in 'zipcode' column: {missing_values}")


# Print all values from the 'zipcode' column
print(edf['zipcode'])

try:
    edf['Latitude'] = nomi.query_postal_code(edf['zipcode'].astype(str)).latitude
except Exception as e:
    print(f"An error occurred: {e}")

edf['Longitude'] = (nomi.query_postal_code(edf['zipcode'].tolist()).longitude)

fig = go.Figure(data=go.Scattergeo(
        lon = edf['Longitude'],
        lat = edf['Latitude'],
        text = edf['zipcode'],
        mode = 'markers',
        marker_color = edf['count'],
        ))

fig.update_layout(
        title = 'colC Distribution',
        geo_scope='usa',
    )
fig.show()
```
